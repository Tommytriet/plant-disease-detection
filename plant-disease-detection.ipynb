#Inference

```python
from PIL import Image
from google.colab import files
import cv2

# Tải lên 10 ảnh từ máy tính của bạn lên Colab
uploaded_files = files.upload()

# Lặp qua các tệp đã tải lên
for i, (file_name, file_data) in enumerate(uploaded_files.items()):
    if i >= 10:
        break  # Chỉ xử lý 10 tệp đầu tiên

    # Lưu tệp vào thư mục hiện tại
    with open(file_name, 'wb') as f:
        f.write(file_data)

    # Đọc ảnh input từ tệp đã tải lên bằng OpenCV
    image = cv2.imread(file_name)

    # Kiểm tra xem ảnh có được đọc thành công không
    if image is not None:
        print(f"Đọc ảnh {file_name} thành công!")
    else:
        print(f"Không thể đọc ảnh {file_name}.")

    image_resized = cv2.resize(image, (1880, 1880))

    # Lưu ảnh đã resize
    resized_file_name = "/content/resized_" + file_name
    cv2.imwrite(resized_file_name, image_resized)

    # Chạy đoạn mã với ảnh đã resize
    !python detect.py --weights "/content/yolov5-master/runs/train/exp/weights/best.pt" --img-size=416 --conf-thres=0.25 --source "{resized_file_name}"
```
